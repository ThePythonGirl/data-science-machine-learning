## Work Flow

```python
#import matplotlib and get it ready for plotting in Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

#setup plot
fig, ax=plt.subplots(figsize=(5, 5)) #(width, height)

#plot data
ax.plot(x, y)

#customize data
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

#save and show
fig.savefig("Untitled Folder/Simple-Plot.png")
```

## Making figures with numpy arrays

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

### Line Plot

```python
#Plot data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

### Scatter Plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### Bar Plot

```python
# Plot using Dictionary
Chocolate_prices = {"Diary milk": 20,
                   "Munch": 25,
                   "Nuts Cracker": 45}
fig, ax = plt.subplots()
ax.bar(Chocolate_prices.keys(), height=Chocolate_prices.values())
ax.set(title="Mady's Chocolate Store", 
      ylabel="Price (Rs)");
```

```python
# Horizontal bar
fig, ax = plt.subplots()
ax.barh(list(Chocolate_prices.keys()), list(Chocolate_prices.values()));
```

## Histogram

```python
x = np.random.randn(1000) #randn - returns samples from "standard normal" distribution
fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots

### Two ways for subplots

```python
#1.
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
#Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(Chocolate_prices.keys(), Chocolate_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# in 1. we plotted to each different axis while in 2. we plotted to each different index - list indexing method
```

```python
#2.
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10, 5))
#Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(Chocolate_prices.keys(), Chocolate_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
