```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
#Use RegEx to replace '$', ',' sign with an empty string 
car_sales["Price"] = car_sales["Price"].str.replace('[\$\, \.]', '')
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
# make new column Sale Date
car_sales["Sale Date"] = pd.date_range('05/01/2020', periods = len(car_sales))
car_sales
```

```python
# to calculate cumsum over price first it needs to converted into int
car_sales["Total Price"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plot total sale - .plot can directly be called on pandas itself
car_sales.plot(x="Sale Date", y="Total Price");
```

```python
# Plot Odometer on scatter
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```

```python
# import another datasets
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Subplots
```

```python
heart_disease.plot.hist(figsize = (10, 20), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# c=target
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10, 6));
```

```python
# OO method with pyplot
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
            x='age',
            y='chol',
            ax=ax);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize the data
ax.set(title="Heart disease and Cholesterol levels",
      xlabel="age",
      ylabel="Cholesterol")
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")
# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# create subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="summer")  # cmap changes the color scheme
# customize ax0
ax0.set(title="Heart disease and Cholesterol levels",
       ylabel="chol")
# change the x limit
ax0.set_xlim([50, 80])
# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="target")
# add a meanline 
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="summer")
# customize a1
ax1.set(title="Heart disease and Max heart rate",
       xlabel="age",
       ylabel="thalach")
# change y limt
ax1.set_ylim([60, 200])
# add legend to ax1
ax1.legend(*scatter.legend_elements(), title="target")
# add a meanline 
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
